```python
# Import library
import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Baca dataset
from google.colab import files
uploaded = files.upload()
```

```python
dataset = pd.read_csv("Statistik_Harian_per_Provinsi_COVID19_Indonesia_Rev.csv")
dataset.tail()
```

```python
# Buat Kolom GMV
dataset['gmv'] = dataset['Kasus_Terkonfirmasi_Akumulatif']
```

```python
# Buat Multi-Line Chart
dataset.groupby(['Tanggal','Provinsi'])['gmv'].sum().unstack().plot()
plt.title('Grafik Harian Kasus Terkonfirmasi Covid 19 per Provinsi',loc='center',pad=30, fontsize=20, color='green')
plt.xlabel('Tanggal', fontsize = 15)
plt.ylabel('Total Amount (in Thousand)',fontsize = 15)
plt.grid(color='gray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000).astype(int))


#Menambah Legend
plt.legend(loc='right', bbox_to_anchor=(1.6, 0.5), shadow=True, ncol=2)
plt.gcf().set_size_inches(20, 10)
plt.tight_layout()
plt.show()
```

```python
plt.scatter(dataset['Kasus_Meninggal_Akumulatif'], dataset['Kasus_Sembuh_Akumulatif'])
plt.title('Data Kasus Meninggal Akumulatif vs Data Kasus Sembuh Akumulatif',loc='center',pad=30, fontsize=20, color='blue')
plt.xlabel('Kasus Meninggal', fontsize = 15)
plt.ylabel('Kasus Sembuh',fontsize = 15)
plt.grid(color='gray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.gcf().set_size_inches(15, 10)
plt.tight_layout()
plt.show()
```
